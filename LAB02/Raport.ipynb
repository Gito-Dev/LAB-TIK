## Hyrje ne Git dhe GitHub
---

Git eshte nje sistem per kontrollin e versioneve (version control system) qe ruan historine e ndryshimeve te skedareve te nje projekti.

Ai perdoret gjeresisht ne:
- zhvillimin e software
- analiza te te dhenave
- projekte kerkimore
- bashkepunim shkencor

GitHub eshte nje platforme online qe perdor Git dhe lejon ruajtjen dhe ndarjen e projekteve ne internet .

Git lejon te:
- ruash historine e punes
- rikthej versionet e meparshme
- bashkpunim me te tjeret ne nje projekt

Pra **Git** ruan historine e punes, ndersa **GitHub** e ruan ate online dhe e ben te akse4

## Instalimi dhe Konfigurimi
---

**Git mund te shkarkohet nga faqja zyrtare:**

https://git-scm.com/

**Pas instalimit kontrollohet me komanden nga terminali:**

git --version

Nese Git eshte instaluar sakte do te shfaqet versioni.

**Pastaj konfiguron identitetin:**

fit config --global user.name "Emri"

git config --global user.email "email@example.com"

**Kontrollon konfigurimin:**

git config --global --list

## Repository
---

Nje **Repository (repo)** eshte nje dosje qe permban projektin dhe historine e tij.

Mund te jete:
- **lokal** (ne kompjuterat tane)
- **online** ne nje platform hosting (GitHub).

Ai permban:
- kodin ose dokumentet
- versionet e meparshme
- versionet e meparshme

## Git Hosting
---
**Git eshte nje mjet lokal, por zakonisht projektet ruhen edhe online.**

Platforma te njohura per **Git hosting** jane:

- GitHub
- GitLab
- Bitbucket

Ne kete kurs ne do te perdorim **GitHub**, nje nga platformat me te perdorura ne bote per projekte software dhe kerkimore.

**GitHub lejon:**

- ruajtjen e projekteve online
- bashkepunim
- dorezimi e puneve laboratorike


## Komandat minimale te Git
---

Ne kete laborator perdorim vetem disa komanda baze:

| Komanda | Kuptimi | 
|---------|---------|
| git init | Krijon nje repository lokal |
| git status | Tregon ndryshimet ne projekt |
| git add. | Shton te gjitha ndryshimet |
| git commit -m "mesazh" | Ruan nje version te ri |
| git push | Ngarkon ndryshimet ne GitHub |
| git pull | Merr ndryshimet nga GitHub |


## Workflow i Git
---
**Workflow i Git** eshte procesi i ruajtjes dhe ngarkimit te ndryshimeve.

Hapat jane :

**1.** Behen ndryshime ne projekt

**2.** Kontrollohet gjendja me git status

**3.** Shtohen ndryshimet me git add.

**4.** Ruhen versioni me git commit

**5.** Ngarkohet ne GitHub me git push

Ky proces perseritet cdo here qe perditesohet projekti.


## Lidhja me GitHub
---
**Per te lidhur projektin lokal me GitHub perdoren keto komanda:**

git remote add origin https://github.com/USERNAME/REPOSITORY.git

git branch -M main

git push -u origin main

Kjo ben qe projekti te lidhet me repository online.
