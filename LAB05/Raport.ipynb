## 2. Ngarkimi dhe analiza e te dhenave :

Te dhenat u lexuan duke perdorur bibliotekën **NumPy**.
Ato perbehen nga dy variabla: **x** dhe **y**, te cilat perfaqesojne marredhenien lineare ne disa variacione te vogla te rastesishme.

Qellimi i analizes eshte të identifikohet marredhenia midis këtyre dy variablave.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("data/model.dat", delimiter=",", skiprows=1)

x_Data = data[:, 0]
y_Data = data[:, 1]
```

## 3. Pershtatja lineare dhe vizualizimi :
Per te modeluar te dhenat u perdor regresioni linear (linear regression), i cili gjen vijen me te mire qe pershtatet me te dhenat.

Grafiku paraqet:
- pikat e te dhenave origjinale
- vijen e pershtatur (best fit line)

Ky model ndihmon ne kuptimin e marredhenies lineare midis variablave x dhe y.

```python
a, b = np.polyfit(x_Data, y_Data, 1)

print(f"Pjerresia e drejtezes a = {a : .2f}")
print(f"Prerja me boshtin b = {b : .2f}")
```

## Vizualizimi i te dhenave dhe modelit :

```python
xfit = np.linspace(1,10,100)
yfit = a*xfit + b

plt.figure(1,(8,6))
plt.plot(x_Data, y_Data, 'ok', label='Te Dhenat')
plt.plot(xfit, yfit, '-r', lw=2, label=f'y={a: .2f}*x + {b: .2f}')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

## Analiza e mbetjeve (residuals) :

```python
plt.figure(2,(8,6))
plt.stem(x_Data, y_Data - (a*x_Data + b), '-r')
plt.hlines(0,0,11,ls='--',color='k',lw=1)
plt.xlabel("x")
plt.ylabel("mbetjet")
plt.xlim(0.8,10.2)
plt.show()
```

## 4. Parametrat e modelit :
Nga pershtatja lineare u perftuan keto vlera:

- a (pjerresia) = 3.08
- b (prerja) = 5.04

Parametri a tregon sa ndryshon y kur x rritet me nje njesi, ndersa b tregon vleren e y kur x = 0.

```python
plt.figure(figsize=(8,6))
plt.stem(x_Data, y_Data - (a*x_Data + b))
plt.hlines(0, 0, 11, linestyle='--')

plt.title("Grafiku i mbetjeve (analiza e gabimeve )")
plt.xlabel("x")
plt.ylabel("mbetjet")

plt.xlim(min(x_Data), max(x_Data))
plt.show()
```

## 5. Analiza e mbetjeve (Residuals) :

Mbetjet perfaqesojne diferencen midis vlerave reale dhe atyre te parashikuara nga modeli linear.

Analiza e mbetjeve eshte shume e rendesishme per te vleresuar cilesine e modelit.

Nese mbetjet shperndahen rastesisht rreth zeros, kjo tregon qe modeli linear eshte i pershtatshem per te dhenat.
