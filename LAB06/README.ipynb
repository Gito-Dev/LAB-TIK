# LAB 05 – Gjenerimi dhe Analiza e te Dhenave (Pershtatje Lineare)
---
## Pershkrimi i projektit :

Ky laborator ka si qellim gjenerimin e nje dataset-i unik duke perdorur skriptin `generate_data.py`, i cili varet nga GitHub ID e studentit.

Pas gjenerimit te te dhenave, ato ruhen ne folderin `data/` në file-in `model.dat`.

Me pas, te dhenat analizohen duke perdorur Python dhe NumPy, ku aplikohet regresioni linear per te gjetur lidhjen midis variablave x dhe y.

Gjithashtu, behet vizualizimi i te dhenave dhe analiza e mbetjeve (residuals) per te vleresuar cilesine e modelit.

## Struktura e projektit :

- **`generate_data.py`** → skripti per gjenerimin e te dhenave
- **`data/model.dat`** → dataset-i i gjeneruar
- **`notebook.ipynb`** → analiza dhe grafikat
- **`Reports/`** → raporti final i laboratorit
